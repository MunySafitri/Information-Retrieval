##Case Folding
*   Case Folding adalah salah satu bentuk text preprosessing yang paling sederhana dan efektif meskipun sering diabaikan
*   Tujuan dari case folding untuk mengubah semua huruf dalam dokumen menjadi huruf kecil.
* hanya huruf 'a' sampai 'z' yang diterima.
* karakter selain huruf dihilangkan dan dianggap delimeter.
* pada tahap ini tidak menggunakan external library apapun, bisa memanfaatkan modul tersedia di python.





##Mengubah text menjadi lowercase
* salah satu contoh pentingnya penggunaaan lower case adalah untuk mesin pencarian.
* bayangkan anda sedang mancari dokumen yang mengandung "indonesia" namun tidak ada hasil yang muncul karena "indonesia" di indeks sebagai "INDONESIA"
* hapuslah angka jika tidak relevan dengan apa yang akan dianalisa, contohnya speerti nomor rumag, nomor telepon, dll.
* Regular Expression (regex) dapat digunakan untuk menghapus karakter angka
* python memiliki module untuk melakukan hal-hal yang berkaitan dengan regex.

##Menghapus tanda baca


*   Sama halnya dengan angka, tanda baca dalam kalimat tidak memiliki perngaruh pada text preprocessing

## Menghapus WhiteSpace
* untuk menghapus spasi di awal dan akhir, dapat menggunakan strip() pada python

````python
import re #modul regular expression
import string
kalimat = """   Ini adalah percobaan teks untuk tugas dua NLP, 
sungguh unik bahasa dan materi ini, jika mendapatkan nilai 100%, sungguh luar biasa   """
lower_case = kalimat.lower()
lower_case = re.sub(r"\d+","",lower_case)#menghilangkan angka
#menghilangkan karakter dan white space
lower_case =lower_case.translate(str.maketrans("","",string.punctuation)).strip()
print(lower_case)
````

````python
#Import word_tokenize dari modul nltk
import nltk

nltk.download("book") 
````

````python
from nltk.tokenize import word_tokenize 
kalimat = kalimat + ". Rumah besar merupakan rumah idaman banyak orang"
tokens = nltk.tokenize.word_tokenize(kalimat)
print(tokens)
````

Kemunculan tanda tanda baca titik(.) dan koma(,), serta token "Rumah" masih menggunakan hutuf besar di awal kata.
* Hal tesebut dapat mengganggu proses perhitungan dalam penerapan algoritma.
Jadi, sebaiknya teks telah melewati tahap case folding seblum di tokenize agar menghasilkan hasil lebih konsisten
* kita dapat juga bisa mendapatkan informasi frekuensi kemunculan setiap token dengan kelas FreqDist() yang sudah tersedia pada modul NLTK,
* untuk menggambarkan ke dalam bentuk grafik, kita perlu menginstall library Matplotlib


````python
from nltk.probability import FreqDist

kalimat = kalimat.translate(str.maketrans('','', string.punctuation)).lower()
tokens = nltk.tokenize.word_tokenize(kalimat)
kemunculan = nltk.FreqDist(tokens)
print(kemunculan.most_common())
````

````python
import matplotlib.pyplot as plt
kemunculan.plot(30, cumulative=False)
plt.show()
````

##fILTERING (STOPWORF REMOVAL)


*   FILTERING ADALAH tahap mengambil kata kata penting dari hasil token dengan menggunakan algoritma stoplist(membuang kata kurang penting) atau wordlist(menyimpan kata penting)
*   Stopword adalah kata umum yang biasanya muncul dalam jumlah besar dan dianggap tidak memilki makna.
* contoh stopword dalam bahasa indonesia adalah "yang", "dan", "di", "dari" dll
* makna di balik penggunaan stopword yaitu dengan menghapus kata-kata yang memilkik informasi rndah dari sebuha teks, fokus pada kata-kata penting.
* contoh penggunaan filtering dapat kita temukan pada konteks mesin pencarian. jika permintaan pencarian adalah "qpq itu pengertian manajemen" tentunga kita ingin sistem pencarian fokus pada memunculkan dokumen dengan topik tentang "pengertian manajemen" diatas dokumen dengan topik "apa itu". Hal ini dilakukan dengan mencegah kata dari daftar stopword dianalisa. 



````python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

print(kalimat)
tokens = word_tokenize(kalimat)
listStopword = set(stopwords.words('indonesian'))

removed = []
for t in tokens:
  if t not in listStopword:
    removed.append(t)

print(removed)
````

##Stemming

##Perbedaan antara Lemmetization dan Stemming
* tujuan dari stemming dan lemmetization adalah untuk mengurangi benyuk infleksi dan kadang-kadang terkait bentuk word ke bentuk dasar yang umum.
* Namun, kedua kata itu berbeda dalam rasanya, Stemming biasanya mengacu pada proses heuristik kasar yang memotong ujung kata-kata dengan harapan mencapai tujuan ini dengan benar sebagian besar waktu, dan sering termasuk menghilangkan imbuhan derivasional.
* Lemmetization biasanya mengacu pada melakukan sesuatu dengan benar dengan menggunakan kosakata dan analisis morfologis kata-kata, biasanya bertujuan untuk menghilangkan ujung inflektif saja dan mengembalikan basis atau bentuk kamus dari suatu kata, yang dikenal sebagai lemma.
* Lemmatisasi terkait erat dengan stemming. perbedaannya adalah bahwa stemmer beroperasi pada satu kata tunggal tanpa pengetahuan konteks, dan oleh karena itu tidak dapat memebedakan anatar kata-kata yang memeiliki arti kata berbeda tergantung pada bagian dari pidato. Namun, stemmer biasanya lebih mudah diimplementasikan dan dijalankan lebih cepat, dan akurasi yang berkurang mungkin tidak masalah untuk bebeapa aplikasi.
*Contohnya:
    - kata "lebih baik" memilki "baik" sebagai lemma-nya. tautan ini tidak tidak terjawab oleh stemming, karena memerlukan penarian kamus. 
    - kata"berjalan" adalah bentuk dasar untuk kata "berjalan" dan karenanya ini cocok dengan stemming dan lemmatisasi
    - kata "meeting" dapat berupa bentuk dasar dari kata benda atau bentuk kata kerja ("to meet") tergantung pada konteksnya misalnya, "dalam pertemuan terakhir kami " atau "kami akan bertemu lagi besok". tidak seperti stemming, lemmatisasi pada prinsipnya dapat memilih lemma yang tepat tergantung pada konteksnya.
* Tujuan dari stemming dan lemmatization adalah untuk mengurangi variasi yang morfologis, ini berbeda dengan prosedur "istilah konflasi" yang lebih umum, yang juga dapat membahas variasi leksico-semantik, sintaksis atau ortografis.
